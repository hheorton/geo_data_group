Demonstrates the new ``shapely.vectorized`` module.

```python
import shapely.vectorized
help(shapely.vectorized)
```

```python
import shapely.geometry as sgeom

import cartopy.io.shapereader as shpreader

shpfilename = shpreader.natural_earth(resolution='10m',
                                      category='cultural',
                                      name='admin_0_countries')
reader = shpreader.Reader(shpfilename)
countries = list(reader.records())

UK, = [country for country in countries if country.attributes['name'] == 'United Kingdom']
UK_geom = UK.geometry
print UK_geom.bounds
```

```python
import numpy as np

y, x = grid_points = np.mgrid[50:59:50j, -8:2:100j]
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs


UK_mask = shapely.vectorized.contains(UK_geom, x, y)

plt.figure(figsize=(12, 12))
ax = plt.axes(projection=ccrs.PlateCarree())
cmap = mcolors.ListedColormap(['blue', 'green'])


ax.scatter(x, y, s=(UK_mask + 0.5) * 20, c=UK_mask + 1, cmap=cmap,
           edgecolor='none', transform=ccrs.PlateCarree())

ax.add_geometries([UK_geom], ccrs.PlateCarree(),
                  facecolor='none', edgecolor='black')

plt.show()
```

To measure performance, let's start with the 20x40 grid case (800 grid points):

```python
%timeit shapely.vectorized.contains(UK_geom, x, y)
```

We can compare this with the analogous pure Python version:

```python
import shapely.prepared

def pure_python_contains(geom, x, y):
    prep = shapely.prepared.prep(geom)
    result = np.empty(x.shape, dtype=np.bool)
    for index in np.ndindex(result.shape):
        result[index] = prep.contains(sgeom.Point(x[index], y[index]))

%timeit pure_python_contains(UK_geom, x, y)
```

So the Cython version is ~10x faster than the pure python version. This speedup factor seems to scale linearly to 80000 grid points:

```python
y, x = grid_points = np.mgrid[50:59:200j, -8:2:400j]

%timeit shapely.vectorized.contains(UK_geom, x, y)
%timeit pure_python_contains(UK_geom, x, y)
```

And 1,000,000 containment tests on a polygon can be computed in under 6s (and quicker than the pure python can compute 80,000):

```python
%timeit shapely.vectorized.contains(UK_geom, *np.mgrid[50:59:1000j, -8:2:1000j][::-1])
```
